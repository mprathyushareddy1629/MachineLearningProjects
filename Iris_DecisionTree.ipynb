```python
import graphviz
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn import tree
```

```python
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
print(iris.DESCR)
iris.target
```

```python
clf.fit(iris.data, iris.target)
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df.corr()
```

```python
def get_decision_tree():
  dot_data = tree.export_graphviz(clf, out_file=None, feature_names = iris.feature_names, class_names=iris.target_names,filled=True, rounded=True, special_characters=True)
  graph = graphviz.Source(dot_data)
  return graph
```

```python
get_decision_tree()
```

### Decision Tree Parameters

```python
clf = tree.DecisionTreeClassifier(max_depth = 2)
clf.fit(iris.data, iris.target)
get_decision_tree()
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy",min_samples_split=30)
clf.fit(iris.data, iris.target)
get_decision_tree()
```

```python
clf = tree.DecisionTreeClassifier(max_depth=4,criterion="entropy",min_samples_split=30)
clf.fit(iris.data, iris.target)
get_decision_tree()
```

```python
clf = tree.DecisionTreeClassifier(max_depth=4,criterion="entropy",min_samples_split=30, min_samples_leaf=20)
clf.fit(iris.data, iris.target)
get_decision_tree()
```
